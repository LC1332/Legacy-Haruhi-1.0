```python
import json
import os
import re

texts = []
for file in os.listdir('Haruhi-Lulu'):
    if file.endswith('.txt'):
        with open(os.path.join("Haruhi-Lulu", file), 'r', encoding='utf-8') as f:
            texts.append(f.read())
```

```python
from copy import deepcopy

index = []
res = []
texts_copy = deepcopy(texts)
for i in range(len(texts_copy)-1):
    if texts_copy[i] not in texts_copy[i+1]:
        res.append(texts_copy[i])

res
```

```python
with open('Haruhi-Lulu_Dialogues.jsonl', 'w+', encoding='utf-8') as f:
    for dialogue in res:
        line_msgs = {}
        values = []
        role_msgs = dialogue.split('\n---\n')
        for msgs in role_msgs:
            if ":" in msgs:
                msg = msgs.split(':')
            elif '：' in msgs:
                msg = msgs.split('：')
            else:
                print(msgs)
                continue
            values.append({"role": msg[0], "text": msg[1][1:-1]})
        line_msgs["dialogue"] = values
        line_msgs["source"] = "user"
        json.dump(line_msgs, f, ensure_ascii=False)
        f.write('\n')
```
