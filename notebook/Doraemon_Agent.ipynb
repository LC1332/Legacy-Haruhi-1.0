````python
from IPython.core.interactiveshell import InputRejected
import json
from datetime import datetime
from datetime import timedelta

def validate_datetime_format(date_str):
    try:
        datetime.strptime(date_str, "%Y-%m-%d %H:%M:%S")
        return True
    except ValueError:
        return False
def check_timetable(json_file, time_str):
    # 读取json文件
    with open(json_file, 'r') as f:
        data = json.load(f)
    # print(data)
    # 将时间字符串转换为datetime对象
    time = datetime.strptime(time_str, '%Y-%m-%d %H:%M:%S')

    # 遍历数据
    for item in data[::-1]:
        start_time = datetime.strptime(item['start_time'], '%Y-%m-%d %H:%M:%S')
        end_time = datetime.strptime(item['end_time'], '%Y-%m-%d %H:%M:%S')
        time_diff = start_time - time
        # 判断时间
        if start_time <= time and time <= end_time:
            print("It is time to " + item['name'] + ". " + item['description'])
        elif time < start_time and time_diff < timedelta(hours=6):
            print(f"距离" + item['name'] + f"开始时间还有{time_diff}")
        elif time > end_time:
            # print(item['description'])
            data.remove(item)
            # print('3')
    print(data)
    # 更新JSON文件
    with open(json_file, 'w') as f:
        json.dump(data, f, indent=4)
    return data
def add_timetable(json_file, time_str):
    # 读取原有的JSON数据
    with open(json_file_path, 'r') as f:
        data = json.load(f)
    new_item_name = input("Please input the name of new item.")
    new_item_description = input("Please input the description of new item.")
    print("Please ensure the start_time of new item is later than " + time_str + ".")
    new_item_start_time = input("Please input the start_time of new item. The format is YYYY-MM-DD hh:mm:ss.")
    while not validate_datetime_format(new_item_start_time):
      new_item_start_time = input("Wrong format. Please input the start_time of new item. The format is YYYY-MM-DD hh:mm:ss.")
    new_item_end_time = input("Please input the end_time of new item. The format is YYYY-MM-DD hh:mm:ss.")
    while not validate_datetime_format(new_item_end_time):
      new_item_end_time = input("Wrong format. Please input the end_time of new item. The format is YYYY-MM-DD hh:mm:ss.")
    # 检验输入格式

    # 创建新的事项对象
    if new_item_start_time <= new_item_end_time:
      new_timetable_item = {
        "name": new_item_name,
        "description": new_item_description,
        "start_time": new_item_start_time,
        'end_time': new_item_end_time
      }
      data.append(new_timetable_item)
    else:
      print("Please ensure the start_time of new item is earlier than end_time")
    # 将新的事项添加到数据列表

    # 保存更新后的JSON数据
    with open(json_file_path, 'w') as f:
        json.dump(data, f)
    return data
def delete_timetable(json_file, time_str):
    # 读取json文件
    with open(json_file, 'r') as f:
        data = json.load(f)
    time = datetime.strptime(time_str, '%Y-%m-%d %H:%M:%S')
    for item in data[::-1]:
        start_time = datetime.strptime(item['start_time'], '%Y-%m-%d %H:%M:%S')
        end_time = datetime.strptime(item['end_time'], '%Y-%m-%d %H:%M:%S')
        time_diff = start_time - time
        # 判断时间
        if time <= end_time:
            print(item)
        elif time > end_time:
            # print(item['description'])
            data.remove(item)
            # print('3')
    del_item_name = input("Please input the name of the item you want to delete.")
    for item in data[::-1]:
      if item['name'] == del_item_name:
        data.remove(item)
    with open(json_file, 'w') as f:
        json.dump(data, f, indent=4)
    return data

# 测试
# time_test = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
# print(time_test)

# process_data('/content/Doraemon_Toolhub/Doraemon_Timetable_Tools.json', datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
# print(data)
````

````python
import csv
import json

json_file_path = '/content/Doraemon_Toolhub/Doraemon_Tools.json'
csv_file_path = '/content/Doraemon_Toolhub/Doraemon_Tools.csv'

def json_to_csv(json_data, csv_file):
    # 解析 JSON 数据
    with open(json_data, 'r') as f:
        data = json.load(f)

    # 定义 CSV 文件的列名和数据列表
    fieldnames = ["name", "description", "image_path", "Tool_action_pair"]
    rows = []

    # 将 JSON 数据转换为适用于 CSV 的数据格式
    for item in data:
        row = {
            "name": item["name"],
            "description": item["description"],
            "image_path": item["image_path"],
            "Tool_action_pair": item["Tool_action_pair"]
        }
        rows.append(row)

    # 写入 CSV 文件
    with open(csv_file, "w", newline="") as f:
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(rows)

json_to_csv(json_file_path, csv_file_path)
````

````python
class QuerytoolsInput(BaseModel):
    """Input for QuarytoolsTool"""

    Query4tools_problem: str = Field(..., description="the action that users want to do or ask for a help or a tool to do")

class QuerytoolsTool(BaseTool):
    name = "Recommand_users_Tools"
    # description = "Used for simple flying. After that, the program should print the image for the path img_path"
    description = 'Useful when users ask Doraemon to recommand tools to solve problems starting with ```Help me, Doraemon!```. Print the response and ask people to choose one from them.'

    def _run(self, Query4tools_problem: str):
        loader = CSVLoader(file_path=csv_file_path)
        docs = loader.load()
        embeddings = OpenAIEmbeddings() #初始化
        '''
        为刚才的文本创建embedding，准备将它们存储在向量存储中，使用向量存储上的from documents方法来实现。
        该方法接受文档列表、嵌入对象，然后我们将创建一个总体向量存储
        '''
        db = DocArrayInMemorySearch.from_documents(docs, embeddings)
        retriever = db.as_retriever() #创建检索器通用接口
        llm_query = ChatOpenAI(temperature = 0.0, max_tokens=1024, model="gpt-3.5-turbo-0613") #导入语言模型
        qdocs = "".join([docs[i].page_content for i in range(len(docs))])  # 将合并文档中的所有页面内容到一个变量中
        response = llm_query.call_as_llm(f"{qdocs} Question: Please list all relevant tools can {Query4tools_problem} \
        in a dictionary with the following structure: "+ '```{"Tool 1": { "name": "", "description": "" }, ...}```')
        print(response)
        print("Please choose a tool from the choice below")
        for meta_tool_key, meta_tool_value in eval(response).items():
            for item_key, item_value in meta_tool_value.items():
                # print(item)
                print(meta_tool_key + "'s " + item_key + ": " + item_value)
        return response

    def _arun(self, problem: str):
        raise NotImplementedError("This tool does not support async")

    args_schema: Optional[Type[BaseModel]] = QuerytoolsInput
````
